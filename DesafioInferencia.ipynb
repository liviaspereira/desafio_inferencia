Crie a variável phq9 - escore Patient Health Questionnaire-9 como a soma das variáveis DPQ010 a DPQ090

```python
  df3["phq9"] = df3[["DPQ010", 
                   "DPQ020", 
                   "DPQ030", 
                   "DPQ040", 
                   "DPQ050", 
                   "DPQ060", 
                   "DPQ070", 
                   "DPQ080", 
                   "DPQ090"]].sum(axis = 'columns', skipna = False)

  df3[["DPQ010", 
     "DPQ020", 
     "DPQ030", 
     "DPQ040",
     "DPQ050", 
     "DPQ060", 
     "DPQ070", 
     "DPQ080", 
     "DPQ090",
     "phq9"]].head()
```

Criar uma variável phq_grp - grupo de sintomas de depressão que assume 
0 (“sem sintomas”) se phq9 < 5, 
1 (“sintomas leves”) se 5 <= phq9 < 10, 
2 (“sintomas moderados”) se 10 <= phq9 < 15, 
3 (“sintomas moderadamente severos”) se 15 <= phq9 < 19 e
4 (“sintomas severos”) se phq9 >= 20

```python
conditions = [
  (df3['phq9'].isna()),
  (df3['phq9'] <= 5),
  (df3['phq9'] > 5) & (df3['phq9'] <= 9),
  (df3['phq9'] > 9) & (df3['phq9'] <= 14),
  (df3['phq9'] > 14) & (df3['phq9'] <= 19),
  (df3['phq9'] > 19)
    ]
values = [np.nan, 0, 1, 2, 3, 4]

df3["phq_grp"] = np.select(conditions, values)

df3["phq_grp2"] = df3["phq_grp"]
```

Novas variáveis a partir da variável phq_grp:

```python
# Agrupando phq_grp em 0 = Sem sintomas, 1 = Sintomas leves, 2 = Sintomas moderados a severos

df3["phq_grp2"] = df3["phq_grp"].replace([3, 4], 2)

df3[["phq_grp2"]].value_counts(sort = False) # Avaliando frequencia
```

Agrupando/recategorizando variáveis qualitativas:

1) Substituindo as categorias 7 e 9 por missing para as variáveis DPQ0X0 e DMDEDUC

2) Agrupando as categorias 2 e 5 na variável RIDRETH1

```python
replace_map = {
  "DPQ010": {7: np.nan, 9: np.nan},
  "DPQ020": {7: np.nan, 9: np.nan},
  "DPQ030": {7: np.nan, 9: np.nan},
  "DPQ040": {7: np.nan, 9: np.nan},
  "DPQ050": {7: np.nan, 9: np.nan},
  "DPQ060": {7: np.nan, 9: np.nan},
  "DPQ070": {7: np.nan, 9: np.nan},
  "DPQ080": {7: np.nan, 9: np.nan},
  "DPQ090": {7: np.nan, 9: np.nan},
  "RIDRETH1": {5: 2}, # Other
  "DMDEDUC": {7: np.nan, 9: np.nan},
  "INDFMINC": {1: np.mean([0,4999]), 2: np.mean([5000,9999]), 3: np.mean([10000,14999]),4: np.mean([15000,19999]),
               5: np.mean([20000,24999]),6: np.mean([25000,34999]), 7: np.mean([35000,44999]), 8: np.mean([45000,54999]),
               9: np.mean([55000,64999]), 10: np.mean([65000,74999]), 11: 75000, 12: np.mean([20000, 90000]), 13: np.mean([0, 19999]),
               77: np.nan, 99: np.nan}
}
df4 = df3.replace(replace_map)
```

Gráfico de barras:

```python
def grafico_barras_prop(data, variable):
    (data[[variable]]
     .value_counts(normalize=True, sort = False)
     .rename("Proportion")
     .reset_index()
     .pipe((sns.barplot, "data"), x=variable, y="Proportion"))
    plt.ylim(0,1)
    plt.show()
```

```python
grafico_barras_prop(df4.replace(label_quali), variable = "RIAGENDR")
```

```python
grafico_barras_prop(df4.replace(label_quali), 
                    variable = "RIDRETH1")
```

# **EDA: Análise Bivariada**

Variável quantitativa versus grupo de depressão

```python
# Função para construir boxplot

def grafico_boxplot_grp(data, variable, label):
    
    if label == "": label = variable
    sns.boxplot(x="phq_grp2", y=variable, data=data)
    plt.ylabel(label)
    plt.show()
```

Perfil com maiores incidências de sintomas de depressão

```python
grafico_boxplot_grp(df4.replace(label_quali), "RIDAGEYR", "Idade")
```
